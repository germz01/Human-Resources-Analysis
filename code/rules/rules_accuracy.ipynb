```
# Funzione per classificare in base alla parte antecedente della rule
```

```
def rule_classify( rule, dataset):

    df=dataset
    antecedent=literal_eval(rule.Antecedent)
    #print antecedent

    predicted_labels=[]

    for i in range(df.shape[0]):

        row=list(df.iloc[i])

        # controllo per ogni item nella parte antecedente della regola se è contenuto nella riga del dataset
        lista_check= [] # questi devono essere tutti veri, affinchè posso classificare come vero
        for item in antecedent:
            lista_check.append( item in set(row))

        check_all=all(lista_check) # controllo che tutti gli item sia almeno una volta dentro la riga
        predicted_labels.append(check_all)
        del(lista_check)
        del(check_all)

    return predicted_labels
    

```

# chiamo la funzione dando la regola e il df come input

```
labels = rule_classify(rule,df)
```
